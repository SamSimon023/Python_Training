#Filter all customers who spend more than 60000 annually.

```python
df.filter(df.annual_spend>60000).select("Customer_id","name").show(30)
```

#Create a new column named "spend_lakh" = annual_spend / 100000.

```python
from pyspark.sql.functions import col
df2=df.withColumn("spend_lakh",col("age")/100000)
df2.show()
```

#Create a new column "customer_type"

```python
from pyspark.sql.functions import when, col

df = df.withColumn("customer_type",
                   when(col("annual_spend") > 70000, "Premium")
                   .otherwise("Standard"))
df.show(30)
```

#Show customers whose name starts with the letter A.

```python
df.filter(df.name.startswith("A")).show()
```

#Filter customers where category is either Clothing or Electronics.

```python
df.filter(df.category.isin("Clothing", "Electronics")).show()
```

#Convert the city name to uppercase using the upper() function.

```python
from pyspark.sql.functions import upper
df.select(upper(df.city)).show(30)
```

#Find customers who are younger than the average age of the entire dataset.

```python
from pyspark.sql.functions import avg

average_age = df.select(avg("age")).collect()[0][0]
filtered_df = df.filter(df.age < average_age)
filtered_df.show()
```

#Extract the first letter of each name and store in a new column.

```python
from pyspark.sql.functions import substring

df = df.withColumn("first_letter", substring(df.name, 1, 1))
df.show()
```

#Mask the customer_id.
Example: C001 â†’ ***01

```python
from pyspark.sql.functions import concat, substring, lit, col

df = df.withColumn("masked_customer_id", concat(lit("***"), substring(col("customer_id"), 3, 2)))
df.select("customer_id", "masked_customer_id").show()
```

#Create a new boolean column "is_senior" where age > 35.

```python
from pyspark.sql.functions import when, col
df = df.withColumn("is_senior", when(col("age") > 35, True).otherwise(False))
df.select("name", "age", "is_senior").show()
```
